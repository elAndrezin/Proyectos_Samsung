```python
ClasesMAX_GP = GP_math['absences'].max()
print(ClasesMAX_GP)

ClasesMAX_MS = MS_math['absences'].max()
print(ClasesMAX_MS)
```

```python
#AsistenciaGP = ((ClasesMAX_GP-GP_math['absences'])/ClasesMAX_GP)*100


AsistMS = ((ClasesMAX_MS - MS_math['absences'])/ClasesMAX_MS) * 100

AsistGP = ((ClasesMAX_GP - GP_math['absences'])/ClasesMAX_GP) * 100

AsistT_Math = AsistGP.add(AsistMS, fill_value = 0)
G3T_Math = GP_math['G3'].add(MS_math['G3'], fill_value = 0)

print(AsistT_Math)

As1 = AsistT_Math.lt(80) 
G3a = G3T_Math.lt(10)
G3b = G3T_Math.lt(15)
i = 0
app = []
ex   = []

while(i < 395): 
    if(As1[i] == True):
        app.append(0)
        ex.append(0)
        i+=1
    elif(As1[i] == False) & (G3a[i] == True):
        app.append(0)
        ex.append(0)
        i+=1
    elif (As1[i] == False) & (G3a[i] == False) & (G3b[i] == True):
        app.append(1)
        ex.append(1)
        i+=1
    elif (As1[i] == False) & (G3b[i] == False):
        app.append(1)
        ex.append(0)
        i+=1 

Dataframe_colegio = Dataframe_colegio.assign(APPROVED = app, EXTRA = ex)
```
